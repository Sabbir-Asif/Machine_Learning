# Linear Regression

Linear regression predicts the relationship between two variables by assuming they have a straight-line connection. It finds the best line that minimizes the differences between predicted and actual values.

## Simple Linear Regression

In a simple linear regression, there is one independent variable and one dependent variable. The model estimates the slope and intercept of the line of best fit, which represents the relationship between the variables. The slope represents the change in the dependent variable for each unit change in the independent variable, while the intercept represents the predicted value of the dependent variable when the independent variable is zero.

The equation that explains linear regression **y = b1x + b0 +e**

Here:<br> 
**b1** is the slope<br>
**b0** is the intercept (where the line crosses Y axis)

**Our goal is to find the best-fit line for the scattered data that has a linear property.**

### Two popular methods to calculate the linear regression
>1. **Least Squares Regression**
>2. **Gradient Descent**


#### Least Squares Regression

Our aim is to calculate the values for slope b1 and intercept b0.
<br>
**y = b1x + b0**

**Steps:**
<br>
1. For each (x,y) point calculate $x^2$ and xy.<br>
2. Calculate sum for all x, y, $x^2$ and xy.<br>
3. Calculate slope b1.

#### Gradient Descent for Linear Regression

>Gradient Descent is one of the optimization algorithms that optimize the cost function (objective function) to reach the optimal minimal solution. To find the optimum solution, we need to reduce the cost function (MSE) for all data points. This is done by updating the values of the slope coefficient (B1) and the constant coefficient (B0) iteratively until we get an optimal solution for the linear function.


In the gradient descent algorithm, the number of steps you’re taking can be considered as the learning rate, and this decides how fast the algorithm converges to the minima.
<br><br>
<img src="https://cdn.analyticsvidhya.com/wp-content/uploads/2024/09/976295.webp">
<br><br>

To update B 0 and B 1, we take gradients from the cost function. To find these gradients, we take partial derivatives for B 0 and B 1.
<br><br>
<img src="https://cdn.analyticsvidhya.com/wp-content/uploads/2024/09/204304-2880x2880.webp" width="600">
<br><br>

We need to minimize the cost function J. One of the ways to achieve this is to apply the batch gradient descent algorithm. In batch gradient descent, the values are updated in each iteration. (The last two equations show the updating of values)

The partial derivates are the gradients, and they are used to update the values of B 0 and B 1. Alpha is the learning rate.



#### Code for linear regression

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7),layout='constrained')
ax.plot(df.area, df.price)
ax.set_xlabel('area')
ax.set_ylabel('price')
plt.show()
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area, df.price, color="red",marker="*")
```

```python
new_df = df.drop('price',axis='columns')
new_df
```

```python
price = df.price
price
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,price)
```

```python
reg.predict([[3300]])
```

```python
area_df = pd.read_csv("areas.csv")
area_df
```

```python
p = reg.predict(area_df)
p
```

```python
area_df['prices'] = p
area_df
```

```python
area_df.to_csv("prediction.csv", index=False)
```
